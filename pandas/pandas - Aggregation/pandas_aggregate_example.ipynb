```python
import pandas as pd
import numpy as np

popularity = {"character":["spiderman", "spiderman","superman", "superman", "ironman", "ironman","ironman","jerry","jerry", "tom", "tom"],
             "popularity":["first","second","third","fourth","fifth","fifth","first","second","third","fourth","fifth"],
             "score1":[60,57,46,59,47,47,60,57,46,59,47],
             "score2":[60,57,46,59,47,47,60,57,46,59,47],
             "gender":['male','male','male','male','male','male','male','female','female','male','male']}


dataframe1 = pd.DataFrame(popularity)
```

```python
char_popularity = dataframe1.groupby(['character','popularity'])
```

```python
list(char_popularity)
```

```python
char_popularity.agg(['mean','max'])
```

```python
def get_mean(value):
    return(np.mean(value))

dataframe2 = char_popularity.agg([get_mean,'max'])
```
